### Common-pool resource simulation 

The following is a simulation of how a community can govern a common resource through a reputation and an identity mechanism. 

**Common-pool resources**

Common-pool resources are publicly available, collectively-managed resources that “add [...] to our sense of belonging.” (Standing, 2019, p.32). The resources that such a system can govern can be of five types: 
- Natural commons – common natural resources like water, land or wind, now often expanded to our planetary climate system; 
- Social commons – community-owned social services like housing, transport services, and healthcare; 
- Civil commons – public institutions of justice that rely on due process and rule-following of all agents involved; 
- Cultural commons –  cultural goods that thrive through collective exchange and enjoyment like arts, sports, media and the infrastructures for them; 
- Knowledge commons – publicly exchanged information, ideas and learning processes.

**Cooperation mechanisms**

There are two relevant mechanisms that can support successful cooperation in community-based resource governance. 
- A reputation-based mechanism, where individual agents develop trust towards each other through conditional reciprocity, leading to cooperation or punishment for defectors. 
- An identity-based mechanism that relies on social norms, which agents learn through observation of others, and which interact with each agent's identification with and affective commitment towards the group. 

**Theoretical hypotheses**


These two parallel mechanisms together constitute a novel cognitive-behavioral framework of CPR governance which yields testable hypotheses about the effects of several key intervention variables on system-level beliefs and behaviors.

These hypotheses are:
1. Increased resource levels should correlate with increased identification and affective commitment as the system moves from an exploitation towards a conservation phase (Mosimane et al., 2012).
2. Directed, incomplete networks should sustain cooperation levels equally as high as complete networks while maintaining more idealistic social norms due to overestimation of others’ appropriation (Shreedhar et al., 2018). 
3. The reputation and identity mechanisms in isolation will result in less cooperation and less sustainable resource use than the combination of the two mechanisms (Bergami & Bagozzi, 2000). 
4. Social interaction should increase the sense of belonging that community members feel and increase identity-based cooperation (Jussila et al., 2012).
5. Cooperation should be higher in groups with homogeneously high levels of identification than in groups with more heterogeneous levels of identification (Habyarimana et al., 2007).
6. In groups without strong social norms, revealing everyone’s current strategy for resource use should increase cooperation (Aflagah et al., 2019).

**Resource system**

Simulating a complex system with ecological, social, economic, and cognitive variables requires making some simplifications and assumptions, which follow from the literature. The model looks at a resource-using community with clear boundaries to its users, as Ostrom (1990) suggests. There are several micro-situational variables that are set to fixed or initial levels. These include:
- Network size, which is bounded at $N = 20$, to implement a community where members know each other and do not have uncertainty about outcomes, as predicted in Aflagah et al. (2019). 
- Resource stock, which starts at level $R = 0$. The resource stock can rise or fall to any real number such that $R \in ℝ$.
- The rate of return $r_i$, denoting the natural increase in the resource per community member at each time period. Theoretically, the rate of return can take on any value for naturally depleting or replenishing resources, such that $r_i \in ℝ$.
- The probability of interaction $p(i)$, which describes the probability that an agent will have an in-depth social interaction with another agent that increases their sense of belonging in the community. In this model, the standard configuration assumes that $p(i) = 0.1$; adjustment to a specific CPR situation requires changes to this variable.

For sake of simplicity, the model assumes that $r_i$ is stable and the resource increases linearly over time. As a result of this assumption, the resource naturally increases at the rate $R(t) = R(t-1)+ N r_i$.

**Agents**

The agents who rely on the resource are connected to some neighboring agents in a network structure. To test the findings by Shreedhar et al. (2018), three types of networks will be tested:
1. A complete network, in which every agent is connected to every other agent and can observe their use. This network has full informational transparency.
2. A Barabasi-Albert network, which constitutes a so-called small world network and most closely resembles real human social ties (Barabasi & Albert, 1999). The network has a power law distribution of degrees, which means that most agents are only connected to a few friends but some high-degree nodes reach across different groups to create an integrated network. Connections are undirected, meaning that linked agents can observe each other.
3. A directed, incomplete network, in the form a random k-out graph, in which agents can observe k other agents who can not necessarily observe them back, so there is less direct reciprocity between agents. 


Following the cognitive-behavioral framework of CPR governance, each agent possesses individual attributes that motivate their behavior and measure their economic and social utilities. The economic utilities surrounding resource use are:
- The agent’s wealth $w_i$, given in terms of resource units. Wealth can rise or fall to any value,  $w_i \in ℝ$. 
- The agents’ resource use capacity $X$. The model makes the assumption that all agents have equal use capacities $X=1$, and this variable is therefore not implemented in the computational simulation because it is not essential to testing the hypotheses.
- The agent’s current use of the resource $u_i(t)$ that results from their beliefs and motivations,
- An agent i’s subjective views of the reputation values $theta_ij$ of other observable agents $j$,
- The sanction cost $c$, which an agent will incur when sanctioning defectors.
- The agents’ needs for a given round $Y$, defined as the basic necessities that the agent requires from the common resource to achieve dignity and advance towards “development” of socially-embedded individual agency, following the work of Sen (2011). Needs are given as a ratio of the agents’ capacity $X$ to use the resource, similarly to the rate of return $r_i$. Agents only succeed at maintaining their own economic well-being, a condition for successful CPR governance, as long as $u_i(t) > Y$ on average over time.

Agents only succeed at maintaining their own economic prosperity, a condition for successful CPR governance, as long as, on average over time, $u_i(t) > Y$. The model assumes that needs are stable across time and homogeneously distributed across the population. As a result of these variable specifications, the use of the resource by the agent at any given time step changes their wealth such that $w_i(t+1) =w_i(t) +  u_i(t) - Y$, where
- $w_i(t+1)$ is the agent’s wealth after the round, and
- $w_i(t)$ is the agent’s previous wealth, initialized at $w_i(t) = 0$ at $t = 0$.

Meanwhile, the underlying agent beliefs that regulate resource use and act as social utilities are:
- The social norms distribution of expected cooperation $ p (c) $, which is a distribution over the random cooperation variable $c \in (0,1)$, and serves to store the expectations for other’s behavior that the agent learns through observation, reputations and identity over time.
- Affective commitment towards cooperation $ A \in [0,1] $, which is the percentage of the social norms expectation that an agent acts on given that emotional attachment is a requirement for turning social norms into action (Mosimane et al., 2012).
- Identification with the community $ Q \in [0,1] $, which is the percentage of overlap between individual and group behaviors and goals, and mediates social norm learning and affective commitment (Jussila et al., 2012).
- The potential loss in identity $L$ of losing the sense of belonging and usefulness of the resource that the community provides. Sanctions by other agents make defectors aware of this loss (Akerlof & Kranton, 2000). 

**Use rules**

The mechanism by which the beliefs produce the agent’s current use $u_i(t)$ of the resource at time step $t$, is simplified into the function: 
$ u_i(t) = 1 - p(c) A_i $. 

The use of the resource, following the cognitive-behavioral framework, depends on 
- the agent’s resource use capacity, $C = 1$, which serves as the upper bound of the fraction of the resource that an agent can use each round.
- the social norms the agent holds which regulate their probability of cooperation $p(c)$, following the work of Xiang et al. (2013), and 
- their potential loss in identity from defecting which is driven by their affective commitment $A_i$, following the work of Jussila et al. (2012).

The use of the resource by the agent at any given time step  ui(t) changes their wealth such that $w_i(t+1) =w_i(t) +  u_i(t) - Y$, where
- $w_i(t+1)$ is the agent’s wealth after the round, 
- $w_i(t)$ is the agent’s previous wealth, initialized at wi(t) = 0 at t = 0, and
- $Y$ are the agent’s needs for the resource.

All agents have a subjective perception of social norms which are updated in a Bayesian learning process with observations of neighbors (Xiang et al., 2013).

The social norms distribution of expected cooperation $p (c)$ derives initially from the agent’s social value orientation $S$, which act as the prior in the Bayesian social norm learning process (Bogaert et al.,2008). In each round thereafter, the agent calculates the likelihood that their current social norms correctly describe the observations of their neighbors, and then they update their social norms accordingly. Crucially, the loss from sanctioning $L$, mediated by the identification $Q_i$, affect the perception of data, such that the perceived cooperation $p(c)' = p(c) + Q_i L$. 

In addition to the norms that agents subscribe to, they also register and update their expectations of the variance around the norm of behaviors, which informs their identification with the group.

The reputation mechanism works through agent's observation of their neighbors and the storage of subjective reputations of those neighbors which can motivate sanctioning of those who defect.

The identity mechanism works through social norm-motivated sanctioning of any perceived defector, and also incorporates the role of social interaction which may boost affective commitment.

## Model Validation

The best way to test whether the included micro-level variables and their interactions are a useful characterization of a CPR situation is to simulate the standard configuration of the model and interpret whether the results confirm the macro-level predictions. 

Following the theoretical hypotheses, the standard configuration that should yield the most cooperation is a community of agents that are:
- connected in a directed, incomplete network, and
- use the combined identity and reputation mechanisms to uphold cooperation within the group.

The community should be able to start with low initial identification and affective commitment – an exploitation phase (Mosimane et al., 2012) – and move towards high levels of both over time.
- The key driver for further cooperation are social norms, which should converge around a common value as agents learn from each other how best to use the resource. 
- Convergence of social norms should reduce the variance and inequity between agents' resource use, bringing up identification, which should, in turn, increase affective commitment (Jussila et al., 2012). 
- As the first hypothesis predicts, resource levels should rise alongside identification and affective commitment as the system moves towards a conservation phase. 

Crucially, the ratio between two micro-situational variables has a large impact on the success of the CPR governance scheme. 
- The first of these variables is the natural rate of return of the resource, 
- the second is the agent need for the resource. 

If natural replenishment or accumulation of the resource is low, it is less likely to meet the needs of the community. 

```python
random_test = CPRSim(network_size=20,mechanism='combined',ID=.2,AC=.2,rate_of_return=0.8,needs=.5,markup=True)
random_test.simulate(50)
```

```python
return_rates = np.linspace(0.7,0.9,5)
wealth_upper_validation = []
wealth_mean_validation = []
wealth_lower_validation = []
resource_upper_validation = []
resource_mean_validation = []
resource_lower_validation = []
ident_upper_validation = []
ident_mean_validation = []
ident_lower_validation = []
aff_comm_upper_validation = []
aff_comm_mean_validation = []
aff_comm_lower_validation = []

for r in return_rates:
    W = []
    R = []
    I = []
    A = []
    for _ in range(10):
        validation_test = CPRSim(network='directed',rate_of_return=r)
        validation_test.simulate(30)
        W.append(validation_test.wealth[-1])
        R.append(validation_test.resource_over_time[-1]) 
        I.append(validation_test.id_over_time[-1])
        A.append(validation_test.ac_over_time[-1])
    wealth_upper_validation.append(np.percentile(W,97.5))
    wealth_mean_validation.append(np.mean(W))
    wealth_lower_validation.append(np.percentile(W,2.5))
    resource_upper_validation.append(np.percentile(R,97.5))
    resource_mean_validation.append(np.mean(R))
    resource_lower_validation.append(np.percentile(R,2.5))
    ident_upper_validation.append(np.percentile(I,97.5))
    ident_mean_validation.append(np.mean(I))
    ident_lower_validation.append(np.percentile(I,2.5))    
    aff_comm_upper_validation.append(np.percentile(A,97.5))
    aff_comm_mean_validation.append(np.mean(A))
    aff_comm_lower_validation.append(np.percentile(A,2.5))
```

```python
fig, axs = plt.subplots(2, 2, sharex=True)
plt.subplots_adjust(top=1.6,right=1.8, hspace=.2,wspace=.2)

axs[0,0].plot(return_rates, wealth_upper_validation,color='grey')
axs[0,0].plot(return_rates, wealth_mean_validation,color='midnightblue')
axs[0,0].plot(return_rates, wealth_lower_validation,color='grey')
axs[0,0].set(title = 'Wealth and mean id in response to return rate',ylabel ='wealth of agents')

axs[0,1].plot(return_rates, resource_upper_validation,color='grey')
axs[0,1].plot(return_rates, resource_mean_validation,color='forestgreen')
axs[0,1].plot(return_rates, resource_lower_validation,color='grey')
axs[0,1].hlines(0,.7,.9,linestyles='dashed')
axs[0,1].set(ylabel ='resource',title = 'Resource levels and mean ac in response to return rate')

axs[1,0].plot(return_rates, ident_upper_validation,color='grey')
axs[1,0].plot(return_rates, ident_mean_validation,color='royalblue')
axs[1,0].plot(return_rates, ident_lower_validation,color='grey')
axs[1,0].set(ylabel='mean id',xlabel='return rate',ylim=(.88,1))

axs[1,1].plot(return_rates, aff_comm_upper_validation,color='grey')
axs[1,1].plot(return_rates, aff_comm_mean_validation,color='firebrick')
axs[1,1].plot(return_rates, aff_comm_lower_validation,color='grey')
axs[1,1].set(ylabel ='mean ac', xlabel='return rate',ylim=(.85,1))

plt.show()
```

A sample simulation of the model for 100 time-steps confirms the predictions of the literature for macro-level behaviors. The standard configuration that follows from the theoretical hypotheses is a community of agents that are connected in a directed, incomplete network, and use the combined identity and reputation mechanisms to uphold cooperation within the group.

## Directed, incomplete networks best support cooperation

I compare the outcomes of a community interacting after 10 time steps with a Barabasi-Albert, a directed incomplete random k-out graph, and a complete graph as network structures. 
The literature predicts that directed, incomplete networks should sustain cooperation at equally as high levels as complete networks while maintaining more idealistic social norms due to overestimation of others’ appropriation. 

### 1. The complete network

In the complete network, all agents are connected to each other and can observe everyone else's cooperation levels.

```python
complete_network = CPRSim(network = 'complete',markup=True)
complete_network.simulate(50)
```

Observations:
- The community where all agents are connected maintains the resource successfully and it rises continuously.
- The complete network leads to varied but very confident social norms in a range between $p(c) \in [0.2,0.5]$
- Wealth falls for all agents, with two groups becoming apparent, one which loses a little, and another that loses a lot. Average sanctions per round are very high, the average agent punishes half the community in every round.
- Identification and affective commitment increase, although a group of agents has lower affective commitment.

However, these are only findings from running one simulation, a series of simulations to find the average outcome will show results that we can be more confident in. Since all four key outcome variables seem to take stable development after 20 time-steps, we will only test until then to be efficient with computational resources.

```python
wealth_complete = []
resource_complete = []
ident_complete = []
aff_comm_complete = []

for _ in range(30):
    complete_test = CPRSim(network='complete')
    complete_test.simulate(20)
    wealth_complete.append(complete_test.wealth[-1])
    resource_complete.append(complete_test.resource_over_time[-1]) 
    ident_complete.append(complete_test.id_over_time[-1])
    aff_comm_complete.append(complete_test.ac_over_time[-1])
```

```python
_max_ = 11
fig, axs = plt.subplots(2, 2)
plt.subplots_adjust(top=1.6,right=1.8, hspace=.2,wspace=.2)

axs[0,0].hist(wealth_complete)
axs[0,0].set(title='Agent wealth and ID, complete network, t=20',xlabel='wealth',ylabel='PDF')
axs[0,0].vlines(np.percentile(wealth_complete,2.5),0,8,color='grey')
axs[0,0].vlines(np.mean(wealth_complete),0,8,color='red')
axs[0,0].vlines(np.percentile(wealth_complete,97.5),0,8,color='grey')

axs[0,1].hist(resource_complete)
axs[0,1].set(title='Resource levels and AC, complete network, t=20',xlabel='resource levels')
axs[0,1].vlines(np.percentile(resource_complete,2.5),0,_max_,color='grey')
axs[0,1].vlines(np.mean(resource_complete),0,_max_,color='red')
axs[0,1].vlines(np.percentile(resource_complete,97.5),0,_max_,color='grey')

axs[1,0].hist(ident_complete,bins=20,range=(0,1))
axs[1,0].set(xlabel='identification',ylabel='PDF')
axs[1,0].vlines(np.percentile(ident_complete,2.5),0,30,color='grey')
axs[1,0].vlines(np.mean(ident_complete),0,30,color='red')
axs[1,0].vlines(np.percentile(ident_complete,97.5),0,30,color='grey')

axs[1,1].hist(aff_comm_complete,bins=20,range=(0,1))
axs[1,1].set(xlabel='affective commitment')
axs[1,1].vlines(np.percentile(aff_comm_complete,2.5),0,15,color='grey')
axs[1,1].vlines(np.mean(aff_comm_complete),0,15,color='red')
axs[1,1].vlines(np.percentile(aff_comm_complete,97.5),0,15,color='grey')
plt.show()

print("Esimates for means after 20 time-steps in a complete network")
print("Wealth:",round(np.mean(wealth_complete),2))
print('CI:[',round(np.percentile(wealth_complete,2.5),2),",",round(np.percentile(wealth_complete,97.5),2),']')
print("Resource level:",round(np.mean(resource_complete),2))
print('CI:[',round(np.percentile(resource_complete,2.5),2),",",round(np.percentile(resource_complete,97.5),2),']')
print("Identification:",round(np.mean(ident_complete),2))
print('CI:[',round(np.percentile(ident_complete,2.5),2),",",round(np.percentile(ident_complete,97.5),2),']')
print("Affective commitment:",round(np.mean(aff_comm_complete),2))
print('CI:[',round(np.percentile(aff_comm_complete,2.5),2),",",round(np.percentile(aff_comm_complete,97.5),2),']')
```

After running 30 simulations and obtaining better estimates for the means, the results from the one simulation hold: 
- Wealth falls significantly in all simulations.
- Resource levels rise significantly in all simulations.
- Identification always stabilizies at 0.98.
- Affective commitment rises in all simulations and balances at around 0.8, with considerable variance.

### 3. The directed, incomplete network

In this network structure, all agents observe five other agents who cannot observe them back. 

```python
directed_network = CPRSim(network = 'directed',markup=True)
directed_network.simulate(50)
```

Observations:
- The directed, incomplete network successfully maintains cooperation at the higher levels than the other two networks.
- Social norms converge in a range around $p(c) \in [0.2,0.4]$ with the notable exception of some outliers who do not cooperate at all. The rest of the community is almost perfectly aligned in their norms, with some variance for extraction allowed in the distributions.
- Most importantly, wealth increases for almost all agents and the average value is significantly higher than for the other network structures, even while the resource is maintained at similarly high values. Sanctions per round are by far the lowest compared to the other two configurations.

## Transparency does not increase cooperation

While findings from Aflagah et al. (2019), which inform hypothesis 6, predict that revealing all agent’s current resource use through increased transparency can increase cooperation, the simulations do not seem to support that implication. Agents’ capacity to have information about more of their community beyond four to five others does not improve cooperation and yields losses in affective commitment. More perfect information can come about either if agents have more time to observe others and register their participation or through technological tools to spread information on behavior across the community. Increasing the number of observed others from one to four increases identification and resource levels. However, thereafter, allowing more transparency does not affect resource levels, and decreases affective commitment and average wealth, while identification stays constant.


```python
neighborhoods = np.linspace(1,16,6,dtype='int')
wealth_upper = []
wealth_mean = []
wealth_lower = []
resource_upper = []
resource_mean = []
resource_lower = []
ident_upper = []
ident_mean = []
ident_lower = []
aff_comm_upper = []
aff_comm_mean = []
aff_comm_lower = []

for n in neighborhoods:
    W = []
    R = []
    I = []
    A = []
    for _ in range(10):
        neighborhood_test = CPRSim(neighborhood_size=n)
        neighborhood_test.simulate(20)
        W.append(neighborhood_test.wealth[-1])
        R.append(neighborhood_test.resource_over_time[-1]) 
        I.append(neighborhood_test.id_over_time[-1])
        A.append(neighborhood_test.ac_over_time[-1])
    wealth_upper.append(np.percentile(W,97.5))
    wealth_mean.append(np.mean(W))
    wealth_lower.append(np.percentile(W,2.5))
    resource_upper.append(np.percentile(R,97.5))
    resource_mean.append(np.mean(R))
    resource_lower.append(np.percentile(R,2.5))
    ident_upper.append(np.percentile(I,97.5))
    ident_mean.append(np.mean(I))
    ident_lower.append(np.percentile(I,2.5))    
    aff_comm_upper.append(np.percentile(A,97.5))
    aff_comm_mean.append(np.mean(A))
    aff_comm_lower.append(np.percentile(A,2.5))
```

## Social interaction increases cooperation through the identity mechanism

As predicted by hypothesis 4 and following the work of Jussila et al. (2012), high-quality social interactions between neighboring agents improve cooperation. An increase in the probability for social interaction from $p(meet) = 0.1$ to $p(meet) = 0.5$ triples the final resource levels, although further increases in social interactions beyond $p(meet) = 0.5 $only promote cooperation marginally.

```python
interaction_probs = np.linspace(.1,.9,3)
wealth_interaction = []
wealth_interaction_means = []
resource_interaction = []
ident_interaction = []
ident_interaction_means = []
aff_comm_interaction = []
aff_comm_interaction_means = []
sanctions_interaction = []

for i in interaction_probs:
    W = []
    Wm = []
    R = []
    I = []
    Im = []
    A = []
    Am = []
    S = []
    for _ in range(30):
        interaction_test = CPRSim(mechanism='identity',p_interaction=i)
        interaction_test.simulate(20)
        W.append(interaction_test.accounts)
        Wm.append(interaction_test.wealth[-1])
        R.append(interaction_test.resource_over_time[-1]) 
        I.append(interaction_test.identification)
        Im.append(interaction_test.id_over_time[-1])
        A.append(interaction_test.affective_commitment)
        Am.append(interaction_test.ac_over_time[-1])
        S.append(interaction_test.sanctions / 400)
    wealth_interaction.append(W)
    wealth_interaction_means.append(Wm)
    resource_interaction.append(R)
    ident_interaction.append(I)
    ident_interaction_means.append(Im)
    aff_comm_interaction.append(A)
    aff_comm_interaction_means.append(Am)
    sanctions_interaction.append(S)
```
